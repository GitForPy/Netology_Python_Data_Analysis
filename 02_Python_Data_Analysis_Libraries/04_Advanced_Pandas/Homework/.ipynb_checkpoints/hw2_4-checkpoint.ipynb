# Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

- если источник traffic_source равен yandex или google, то в source_type - ставится organic
- для источников paid и email из России - ставим ad
- для источников paid и email не из России - ставим other
- все остальные варианты берем из traffic_source без изменений

```python
import pandas as pd
```

```python
df = pd.read_csv('visit_log.csv', sep=';')
df.head()
```

```python
%time 

# v.1
df.loc[df.traffic_source.isin(['yandex', 'google']), 'source_type'] = 'organic'
df.loc[ (df.traffic_source.isin(['paid', 'email'])) & (df.region == 'Russia'), 'source_type'] = 'ad'
df.loc[ (df.traffic_source.isin(['paid', 'email'])) & (df.region != 'Russia'), 'source_type'] = 'other'

df.source_type.fillna(df.traffic_source, inplace=True) # Можно было и так: log.loc[log.source_type.isna(), 'source_type'] = log.traffic_source
df.head()
```

```python
# v.2
def creat_source_type(row):
    traffic = row['traffic_source']
    region = row['region']

    if traffic in ['yandex', 'google']:
        row['source_type'] = 'organic'

    if traffic in ['paid', 'email']:
        if region == 'Russia':
            row['source_type'] = 'ad'
        else:
            row['source_type'] = 'other'
    return row
```

```python
%time
df = df.apply(creat_source_type, axis=1)
df.head()
```

Вывод: первый способ работает быстрей, предпочтительней использовать конструкцию с df.loc[ ]

Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
# v.1 - решение через оконные функции:

def window_func_count(df):
  df['counter'] = df['timestamp'].count()
  df['range'] = df['timestamp'].max() - df['timestamp'].min()

  return df
```

```python
# На уровне каждого пользователя рассчитываю кол-во отзывов и lifetime

data = df.groupby('userId').apply(window_func_count)
data.head()
```

```python
# Оставляю только тех пользователей, у которых более 100 оценок.

active_user = data.loc[data.counter > 100]
active_user.head()
```

```python
# Рассчитываю среднее время жизни пользователей:

life_time = active_user.groupby('userId').min()[ ['range']]
life_time.head()
```

```python
# v.2 - Решение с помощью джоинов.

# Формирую список пользователей, у которых более 100 оценок

count_ratings = df.groupby('userId').count()['timestamp'].reset_index()
count_ratings.rename(columns={'timestamp':'counter'}, inplace=True)
count_ratings = count_ratings.loc[count_ratings.counter > 100]

count_ratings.head()
```

```python
# Считаю для каждого пользователя lifetime - разница между максимальным и минимальным значением столбца timestamp

def user_lifetime(df):
    return df['timestamp'].max() - df['timestamp'].min()
```

```python
users_lt = df.groupby('userId').apply(user_lifetime).reset_index()
users_lt.rename(columns={0:'lifetime'}, inplace=True)
users_lt.head()
```

```python
# Склеиваю два датафрейма в один и считаю среднее время жизни пользователей, которые выставили более 100 оценок
output = count_ratings.merge(users_lt, on='userId', how='left')
output.head()
```

```python
mean_lt = output['lifetime'].mean()
print(f'Среднее время жизни пользователей, которые выставили более 100 оценок {int(mean_lt)} сек, что составляет {round(mean_lt / 86400, 2)} дн.')
```

# Задание 4

Дана статистика услуг перевозок клиентов компании по типам:
- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)

auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)

air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)

client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1',
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
```

Нужно сформировать две таблицы:

- таблицу с тремя типами выручки для каждого client_id без указания адреса клиента;
- аналогичную таблицу по типам выручки с указанием адреса клиента.

```python
# Объединяю данные по перевозкам по 3 видам: rzd, auto и air

df_in_concated = pd.concat([rzd, auto, air])

# На случай, если в базе появятся дубликаты:
df_in_concated = df_in_concated.drop_duplicates()

df_in_concated
```

```python
# Формирую сводные данные на уровне client_id:

revenue_all_category = df_in_concated.groupby('client_id').sum().reset_index()
revenue_all_category
```

```python
# Присоединяю к справочнику клиентов информацию о перевозках.
# Использую left join чтобы не потерять клиентов, у которых нет перевозок.
# Также можно было использовать outer на случай, если в базе ошибочно заведены транзакции, которых нет в client_base.

data = client_base.merge(revenue_all_category, on='client_id', how='left')
data
```
