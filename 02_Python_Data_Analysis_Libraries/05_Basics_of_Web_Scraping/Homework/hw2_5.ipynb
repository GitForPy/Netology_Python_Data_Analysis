# Дополнительная часть

Дополнительная часть (необязательная)
Функция из обязательной части задания должна быть расширена следующим образом:

- кроме списка ключевых слов для поиска необходимо объявить параметр с количеством страниц поисковой выдачи. Т.е. при передаче в функцию аргумента 4 необходимо получить материалы с первых 4 страниц результатов;
- в датафрейме должны быть столбцы с полным текстом найденных материалов и количеством лайков:
<дата> - <заголовок> - <ссылка на материал> - <текст материала> - <количество лайков>

```python
import time

def get_posts_from_habr(words, pages=1):

    """
     Функция в качестве параметров принимает:
         - список ключевых слов для поиска (например, ['python', 'анализ данных'])
         - количество страниц поисковой выдачи, т.е.при передаче в функцию аргумента 4 функция возвращает
           материалы с первых 4 страниц результатов поиска.

    На основе материалов, попавших в результаты поиска по каждому запросу, возвращается датафрейм вида:
         <дата> - <заголовок> - <ссылка на материал> - <текст материала> - <количество лайков> - <Страница>

    """

    habr_blog = pd.DataFrame()
    for page in range(pages):
        url = f'https://habr.com/ru/search/page{page + 1}'

        for word in words:
          params = {
                    'q': word,
                    'target_type':'posts',
                    'order':'relevance'
          }
          res = requests.get(url, params)
          time.sleep(0.33)
          soup = BeautifulSoup(res.text)
          query = word.capitalize()

          articles = soup.find_all('article', class_="tm-articles-list__item")

          for article in articles:
              if  article.find('time').text:
                  date = article.find('time').text
              else:
                  date = 'Не указана'

              if article.find('h2').text:
                  title = article.find('h2').text
              else:
                  title = 'Без заголовка'

              if article.find('a', class_= 'tm-article-snippet__title-link'):
                  link = article.find('a', class_= 'tm-article-snippet__title-link').get('href')
                  full_link = 'https://habr.com' + link
              else:
                  link = 'Без ссылки'

              row = [{'query':query,'date':date,'title':title, 'link':full_link, 'page':page + 1}]

              habr_blog = pd.concat([habr_blog, pd.DataFrame(row)]).reset_index(drop=True)

  #     Из ссылок вытаскиваю текст новостей и записываем в отдельное поле в датафрейме full_text:
  #     Также записываю в отдельное поле кол-во лайков.

        for i, post in enumerate(habr_blog['link'].copy()):
            print(post)
            req = requests.get(post)
            time.sleep(0.3)
            soup = BeautifulSoup(req.text)

    #         Добавил еще общий рейтинг
            rating = soup.find('div', class_='tm-votes-meter tm-article-rating__votes-switcher').text

            likes = soup.find('div', class_='tm-votes-meter tm-article-rating__votes-switcher').find('span').text
            full_text = soup.find('div', {'id':"post-content-body"}).text.strip()
            habr_blog.loc[i, 'rating'] = rating
            habr_blog.loc[i, 'likes'] = likes
            habr_blog.loc[i, 'full_text'] = full_text

        habr_blog.to_excel('habr.xlsx')
        return habr_blog

  # data_table.disable_dataframe_formatter()

get_posts_from_habr(['python', 'анализ данных'], 2)
```
